```python
import numpy as np
# Beispiel: Kleine Datenanalyse in Jupyter

import pandas as pd
import matplotlib.pyplot as plt

# 1. Mini-Datensatz erstellen
data = {
    "Name": ["Anna", "Ben", "Clara", "David"],
    "Punkte": [85, 92, 78, 90]
}
df = pd.DataFrame(data)

# 2. Tabelle anzeigen
df

# 3. Einfaches Diagramm zeichnen
df.plot(x="Name", y="Punkte", kind="bar", legend=False)
plt.ylabel("Punkte")
plt.title("Test-Ergebnisse")
plt.show()
```
